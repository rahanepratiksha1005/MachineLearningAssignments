```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
from sklearn.preprocessing import LabelEncoder

# Load dataset
purchase = pd.read_csv("Purchase_new.csv")

# Encode categorical columns
le = LabelEncoder()
for col in ['Holiday', 'Discount', 'Free Delivery', 'Purchase']:
    purchase[col] = le.fit_transform(purchase[col])

# Split dataset into features and target variable
feature_cols = ['Holiday', 'Discount', 'Free Delivery']
X = purchase[feature_cols]
y = purchase['Purchase']

# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

# Create Decision Tree classifier object
clf = DecisionTreeClassifier()

# Train Decision Tree Classifier
clf = clf.fit(X_train, y_train)

# Predict the response for test dataset
y_pred = clf.predict(X_test)

print("Predictions:", y_pred)
```

```python
# Model Accuracy, how often is the classifier correct?
print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

# Step 1: Initialize the Decision Tree
dtc = DecisionTreeClassifier(random_state=42)

# Step 2: Train the Decision Tree on your training data
dtc.fit(X_train, y_train)

# Step 3: Plot the trained decision tree
plt.figure(figsize=(25,12))  # Adjust size for readability
plot_tree(
    dtc,
    feature_names=X_train.columns,  
    class_names=['No Purchase', 'Purchase'],  # target classes
    filled=True,
    rounded=True,
    fontsize=10
)
plt.title("Decision Tree for Purchase Dataset")
plt.show()



```
